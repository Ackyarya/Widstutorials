# Matplotlib 2_1: Adding Subplots and Object-oriented approach

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
def f(t):
    """Python function to calculate a enclosed sinusoid"""
    val = 1 / ((t-2.5)**2 + 1) * np.cos(4*np.pi*t)
    return val

t = np.arange(0.0, 5.0, 0.02)
```

```python
# From what we know till now
plt.plot(t, f(t), 'k')
plt.plot(t, 50*np.cos(2*np.pi*t), 'r--')
```

```python
#Working with multiple figures and axes
plt.subplot(2, 1, 1) # (2, 1, 1) implies first axes in a grid of 2x1 (2 rows, 1 col)
plt.plot(t, f(t), 'k') # plotting in first axes because it is the current axes

plt.subplot(2, 1, 2) # (2, 1, 2) implies second axes in a grid of 2x1 (2 rows, 1 col)
plt.plot(t, 50*np.cos(2*np.pi*t), 'r--') # plotting in second axes because it is the current axes
```

```python
plt.subplot(2, 2, 1) # 1st element of 2x2 grid - 1st row, 1st col
plt.plot(t, f(t), 'b') # Blue graph 

plt.subplot(2, 2, 2) # 2nd element of 2x2 grid - 1st row, 2nd col
plt.plot(t, f(t), 'g') # Green graph

plt.subplot(2, 2, 3) # 3rd element of 2x2 grid - 2nd row, 1st col
plt.plot(t, f(t), 'r') # Red graph

plt.subplot(2, 2, 4) # 4th element of 2x2 grid - 2nd row, 2nd col
plt.plot(t, f(t), 'k') # Black graph
```

```python
#add_subplot() - Substitute of subplot()
plt.subplot(1, 1, 1)
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.ylim(5, 0) # To invert y-axis
```

```python
fig = plt.figure()  # create a figure object
ax = fig.add_subplot(1, 1, 1)  # create an axes object in the figure
ax.plot([1, 2, 3, 4])
ax.set_ylabel('some numbers')
ax.set_ylim(5, 0) # To invert y-axis
```

```python
fig1 = plt.figure()
fig2 = plt.figure() # Defined all the figures simultaneously

ax1 = fig1.add_subplot(1, 1, 1)
ax2 = fig2.add_subplot(2, 1, 1)
ax3 = fig2.add_subplot(2, 1, 2)

ax1.plot(t, f(t), 'b') # Blue Graph
ax2.plot(t, f(t), 'g') # Green Graph
ax3.plot(t, f(t), 'r') # Red Graph
```

```python
#subplots()
fig, axes = plt.subplots(nrows=2, ncols=2) # axes is the multidimensional numpy array of Axes.

axes[0,0].plot(t, f(t), 'b') # Blue graph
axes[1,0].plot(t, f(t), 'g') # Green graph
axes[0,1].plot(t, f(t), 'r') # Red graph
axes[1,1].plot(t, f(t), 'k') # Black graph

# To iterate over all items (along rows) in a multidimensional numpy array, use the `flat` attribute
for ax in axes.flat:
    ax.set(xlim=[0, 3])
```

```python
fig1, axes1 = plt.subplots(2, 2)

# Share a X axis with each column of subplots
fig2, axes2 = plt.subplots(2, 2, sharex='col')

# Share a Y axis with each row of subplots
fig3, axes3 = plt.subplots(2, 2, sharey='row')

# Share both X and Y axes with all subplots
fig4, axes4 = plt.subplots(2, 2, sharex='all', sharey='all')

# Note that this is the same as
fig5, axes5 = plt.subplots(2, 2, sharex=True, sharey=True)

# Setting title of figure is using suptitle()
fig1.suptitle('1st')
fig2.suptitle('2nd')
fig3.suptitle('3rd')
fig4.suptitle('4th')
fig5.suptitle('5th')
```

```python
fig = plt.figure(constrained_layout=True)
# Remove the constrained_layout part and see the difference.

gs = fig.add_gridspec(2, 3) # Creates 6 elements in form of 2x3 grid
# If the above line is showing an error then you must have Matplotlib version 2.2.
# Either update Matplotlib version to 3.x or uncomment the following code.
# import matplotlib.gridspec as gridspec
# gs = gridspec.GridSpec(2, 3)

# Merges 1st element(0,0), 2nd element(0,1)
ax2 = fig.add_subplot(gs[0, :-1])
# [0, :-1] implies 1st row and all columns before last (Recall slicing of list)
ax2.set_title('gs[0, :-1]')

# Merges 3rd element(0,2), 6th element(1,2)
ax3 = fig.add_subplot(gs[:, -1])
ax3.set_title('gs[:, -1]')

# Only 4th element(1,0)
ax4 = fig.add_subplot(gs[1, 0])
ax4.set_title('gs[1, 0]')

# Only 5th element(1,1)
ax5 = fig.add_subplot(gs[1, -2])
ax5.set_title('gs[1, -2]')
```

```python
fig = plt.figure(figsize=(8,6), constrained_layout=True)
widths = [2, 3, 1.5]
heights = [1, 3, 2]
spec = fig.add_gridspec(ncols=3, nrows=3, width_ratios=widths, height_ratios=heights)
for row in range(3):
    for col in range(3):
        ax = fig.add_subplot(spec[row, col])
        ax.set(title=str(row*3 + col))
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# %matplotlib notebook
```

```python
path='NGC5272.csv'
F_275 = np.loadtxt(path, usecols=1)
F_336 = np.loadtxt(path, usecols=2)
F_438 = np.loadtxt(path, usecols=3)
F_606 = np.loadtxt(path, usecols=4)
F_814 = np.loadtxt(path, usecols=5)
prob = np.loadtxt(path, unpack=True, usecols=6)
```

```python
F275, F336, F438, F606, F814 = [[] for i in range(5)]
count = 0
for index in range(len(prob)):
    if prob[index]>90:
        count += 1
        F275.append(F_275[index])
        F336.append(F_336[index])
        F438.append(F_438[index])
        F606.append(F_606[index])
        F814.append(F_814[index])
```

```python
def scplot(axes, mag1, mag2):
    color = [mag1[i] - mag2[i] for i in range(count)]
    axes.scatter(color, mag1, s=0.05)
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(12,8)) # We create the 2x2 grid

# We choose limits such that neither of the subplots is cut.
axes[1,0].set_xlim(-2.0, 5)
axes[1,0].set_ylim(30, 11) # Don't forget to invert the y-axis

# We give appropriate axes labels
axes[1,0].set_xlabel('Color')
axes[1,1].set_xlabel('Color')
axes[1,0].set_ylabel('Apparent Magnitude')
axes[0,0].set_ylabel('Apparent Magnitude')

# We give appropriate subplot labels
axes[0,0].set_title('F275 Vs F275-F336')
axes[0,1].set_title('F336 Vs F336-F438')
axes[1,0].set_title('F438 Vs F438-F606')
axes[1,1].set_title('F606 Vs F606-F814')

# Now, we call scplot() to plot.
scplot(axes[0,0], F275, F336)
scplot(axes[0,1], F336, F438)
scplot(axes[1,0], F438, F606)
scplot(axes[1,1], F606, F814)
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_axes([0.1, 0.6, 0.4, 0.35], ylim=(30,14), xlim=(-2.0, 5), xlabel='F275-F336', ylabel='F275')
ax2 = fig.add_axes([0.6, 0.6, 0.4, 0.35], ylim=(28,14), xlim=(-2.0, 3.4), xlabel='F336-F438', ylabel='F336')
ax3 = fig.add_axes([0.1, 0.1, 0.4, 0.35], ylim=(28,13), xlim=(-0.7, 2.5), xlabel='F438-F606', ylabel='F438')
ax4 = fig.add_axes([0.6, 0.1, 0.4, 0.35], ylim=(25,11), xlim=(-0.4, 2), xlabel='F606-F814', ylabel='F606')
# This shows another advantage of add_axes() over subplots()

ax1.set_title('F275 Vs F275-F336')
ax2.set_title('F336 Vs F336-F438')
ax3.set_title('F438 Vs F438-F606')
ax4.set_title('F606 Vs F606-F814')

scplot(ax1, F275, F336)
scplot(ax2, F336, F438)
scplot(ax3, F438, F606)
scplot(ax4, F606, F814)
```
